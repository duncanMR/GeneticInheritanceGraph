```python
import numpy as np
import tskit
import itertools
import networkx as nx
from ipywidgets import IntSlider, HBox, VBox, Output, Button
from IPython.display import display, SVG, clear_output
import matplotlib.pyplot as plt


def translate_coordinate(edge, x, direction):
    if direction == "up":
        if edge.child_left <= x < edge.child_right:
            parent_length = edge.parent_right - edge.parent_left
            child_length = edge.child_right - edge.child_left
            if parent_length == -child_length:
                # inversion
                return edge.parent_left - (x - edge.child_left)
            elif edge.child_left != edge.parent_left:
                # shift
                return x - (edge.child_right - edge.parent_right)
            else:
                # no change
                return x

    elif direction == "down":
        if (
            edge.parent_left <= x < edge.parent_right
            or edge.parent_left > x >= edge.parent_right
        ):
            parent_length = edge.parent_right - edge.parent_left
            child_length = edge.child_right - edge.child_left
            if parent_length == -child_length:
                # inversion
                return edge.child_left + (edge.parent_left - x)
            elif edge.child_left != edge.parent_left:
                # shift
                return x + (edge.child_right - edge.parent_right)
            else:
                # no change
                return x
    else:
        raise ValueError(f"Invalid direction {direction}")




class GIG:
    """
    A GIG is an ARG in which edges are annotated with coordinate
    transform functions. Nodes in a GIG are integers, and represent
    specific ancestral genomes (identical to an ARG). Edges
    represent a genetic inheritance relationship between two
    genomes. Finally, coordinate transforms define what parts
    of a genome were inherited and the effects of any structural
    variants that may have occured.
    """

    def __init__(self, table_group):
        # TODO instantiate the model by reading in some data and filling
        # in the instance variables as required to define the operations
        # below
        self.intervals = table_group.intervals._df()
        self.nodes = table_group.nodes._df()
        
        pass

    @property
    def num_nodes(self):
        """
        Return the number of nodes in the GIG
        """
        return len(self.nodes)

    def parents(self, u):
        """
        Return the parents of node u in the graph.
        """
        all_parents = self.intervals[self.intervals.child == u]["parent"]
        return all_parents.drop_duplicates()

    def children(self, u):
        """
        Return the children of node u in the graph.
        """
        all_children = self.intervals[self.intervals.parent == u]["child"]
        return all_children.drop_duplicates()

    def local_tree_tskit(self, u, x):
        """
        Returns a tskit tree sequence object based on the local tree for node
        u at position x.
        """
        pi, omega = self.local_tree_bottom_up(u, x)
        tables = tskit.TableCollection(sequence_length=1)
        node_table = tables.nodes
        edge_table = tables.edges
        mutation_table = tables.mutations
        sites_table = tables.sites
        node_table.metadata_schema = tskit.MetadataSchema.permissive_json()
        mutation_table.metadata_schema = tskit.MetadataSchema.permissive_json()

        sites_table.add_row(
            position=0,
            ancestral_state="0",
        )

        for index, row in self.nodes.iterrows():
            omega_val = omega[index]
            node_table.add_row(
                time=row["time"],
                flags=row["flags"],
                individual=row["individual"],
                metadata={"omega": list(omega_val)},
            )
            if omega_val != set():
                mutation_table.add_row(
                    site=0,
                    node=index,
                    derived_state="1",
                    parent=-1,
                    metadata={"omega": list(omega_val)},
                )

        for child, parent in enumerate(pi):
            if parent != -1:
                edge_table.add_row(left=0, right=1, parent=parent, child=child)
        tables.sort()
        ts = tables.tree_sequence()
        return ts

    def plot_local_tree(self, u, x, size=(400, 400)):
        ts = self.local_tree_tskit(u=u, x=x)
        mut_labels = {}
        for mut in ts.mutations():
            mut_labels[mut.id] = ",".join(str(val) for val in mut.metadata["omega"])
        style = (
            ".node > .lab {font-size: 18px; font-family: Arial}"
            ".mut > .lab {font-size: 14px; font-family: Arial}"
            f".n{u} .sym {{fill: blue}}"
            f".node.n{u} > .lab {{fill: blue}}"
        )

        return ts.first().draw_svg(mutation_labels=mut_labels, size=size, style=style)

    def coordinate_transform(self, p, c, direction):
        """
        Return the coordinate transform function between the pair
        of nodes u and v. A coordinate transform function for nodes u and v
        is a function f(x) mapping a positive floating point number in the coordinate
        space of u to a (possibly empty) set of positive reals in the coordinate
        space of v.

        Note that the pair of nodes (u, v) is *directed*. For given pair of nodes
        child and parent nodes (c, p), the coordinate transform for c->p is a
        different function to p->c.
        """
        matching_edges = self.intervals.query("parent == @p and child == @c")

        def transform(x, tables, pi, callback=None, capture=False):
            x_vals = []
            tab = tables.copy()
            for _, edge in matching_edges.iterrows():
                x_val = translate_coordinate(edge, x, direction)
                if x_val is not None:
                    x_vals.append(x_val)
                    if capture is True:
                        if pi[c] == -1:
                            print(f"adding mutation above {c}")
                            tab.edges.add_row(left=0, right=1, parent=p, child=c)
                            tab.mutations.add_row(
                                site=0,
                                node=c,
                                derived_state="1",
                                parent=-1,
                                metadata={"omega": []},
                            )
                            pi[c]=p
                        omega = set(tab.nodes[c].metadata["omega"])
                        omega = omega.union(set([x_val]))
                        tab.nodes[c] = tab.nodes[c].replace(
                            metadata={"omega": list(omega)}
                        )
                        
                        print(tab.mutations)
                        mut_index = np.where(tab.mutations.node == c)[0][0]
                        tab.mutations[mut_index] = tab.mutations[mut_index].replace(
                            metadata={"omega": list(omega)}
                        )

                    if callback is not None:
                        callback(
                            edge=edge,
                            matching_edges=matching_edges,
                            direction=direction,
                            x_old=x,
                            x_new=x_val,
                            tables=tab,
                        )
            return set(x_vals), tab

        return transform


    
    def local_tree_top_down(self, u, x, callback=None):
        """
        Return the local tree for node u at position x in its coordinate space
        as a tuple (pi, omega). Pi is an oriented tree, such that pi[v] is the
        parent of node v, and pi[v] = -1 if u is a root. Omega defines the
        the set of genome coordinate corresponding to x in the coordinate space
        of v for each node v in the tree.
        """

        pi = np.full(self.num_nodes, -1, dtype=int)
        omega = [set() for _ in range(self.num_nodes)]

        # Initialise tskit
        tables = tskit.TableCollection(sequence_length=1)
        tables.nodes.metadata_schema = tskit.MetadataSchema.permissive_json()
        tables.mutations.metadata_schema = tskit.MetadataSchema.permissive_json()
        tables.sites.add_row(
            position=0,
            ancestral_state="0",
        )
        for _, row in self.nodes.iterrows():
            tables.nodes.add_row(
                time=row["time"],
                flags=row["flags"],
                individual=row["individual"],
                metadata={"omega": []},
            )

        at_root = False
        c = u
        while not at_root:
            at_root = True
            for p in self.parents(c):
                f = self.coordinate_transform(p=p, c=c, direction="up")
                X, tables = f(x, tables=tables, pi=pi, callback=callback, capture=False)
                assert len(X) in [0, 1]
                if len(X) == 1:
                    x = X.pop()
                    c = p
                    at_root = False
                    break

        # add root to tree
        tables.mutations.add_row(
            site=0,
            node=c,
            derived_state="1",
            parent=-1,
            metadata={"omega": [x]},
        )
        tables.nodes[c] = tables.nodes[c].replace(metadata={"omega": [x]})
        omega[-1] = set([x])

        # Traverse downwards from the root, updating the tree as we go.
        stack = [(c, x)]
        while len(stack) > 0:
            p, x = stack.pop()
            for c in self.children(p):
                f = self.coordinate_transform(p=p, c=c, direction="down")
                X, tables = f(x, tables=tables, pi=pi, callback=callback, capture=True)
                omega[c] = omega[c].union(X)
                for x_val in omega[c]:
                    pi[c] = p
                    stack.append((c, x_val))
        return pi, omega

    def local_tree_bottom_up(self, u, x, callback=None):
        """
        Return the local tree for node u at position x in its coordinate space
        as a tuple (pi, omega). Pi is an oriented tree, such that pi[v] is the
        parent of node v, and pi[v] = -1 if u is a root. Omega defines the
        the set of genome coordinate corresponding to x in the coordinate space
        of v for each node v in the tree.
        """

        pi = np.full(self.num_nodes, -1, dtype=int)
        omega = [set() for _ in range(self.num_nodes)]

        # Initialise tskit
        tables = tskit.TableCollection(sequence_length=1)
        tables.nodes.metadata_schema = tskit.MetadataSchema.permissive_json()
        tables.mutations.metadata_schema = tskit.MetadataSchema.permissive_json()
        tables.sites.add_row(
            position=0,
            ancestral_state="0",
        )
        for _, row in self.nodes.iterrows():
            tables.nodes.add_row(
                time=row["time"],
                flags=row["flags"],
                individual=row["individual"],
                metadata={"omega": []},
            )

        # Recursive helper function for upward traversal
        def traverse_upward(c, x, tables):
            #print(f'Going up from node {c} at x = {x}')
            parents = self.parents(c)
            if len(parents) == 0:
                return
            for p in parents:
                f = self.coordinate_transform(p=p, c=c, direction="up")
                X, tables = f(x, tables=tables, pi=pi, callback=callback, capture=True)
                assert len(X) in [0, 1]
                if len(X) == 1:
                    x = X.pop()
                    # print(f'setting pi[{c}] = {p}')
                    pi[c] = p
                    omega[c] = omega[c].union(X)
                    traverse_upward(p, x, tables)
                    traverse_downward(p, x, tables)  # Traversing sibling subtrees
                    return

        # Function for downward traversal of sibling subtrees
        def traverse_downward(p, x, tables):
            #print(f'Going down from node {p} at x = {x}')
            stack = [(p, x)]
            while len(stack) > 0:
                p, x = stack.pop()
                for c in self.children(p):
                    f = self.coordinate_transform(p=p, c=c, direction="down")
                    X, tables = f(x, tables=tables, pi=pi, callback=callback, capture=True)
                    omega[c] = omega[c].union(X)
                    for x_val in omega[c]:
                        pi[c] = p
                        stack.append((c, x_val))

        # Start the recursive traversal
        traverse_upward(u, x, tables)

        return pi, omega



    def plot_gig(self, size=(10, 10), fontsize=8):
        """
        Plot the entire GIG using networkx
        """
        nodes = self.nodes
        edges = self.intervals
        edges["parent_interval"] = [
            f"[{e.parent_left}, {e.parent_right})" for e in edges.itertuples()
        ]
        edges["child_interval"] = [
            f"[{e.child_left}, {e.child_right})" for e in edges.itertuples()
        ]
        edges["id"] = range(len(edges))

        # Create directed graph
        G = nx.from_pandas_edgelist(
            edges,
            source="parent",
            target="child",
            edge_attr=True,
            create_using=nx.MultiDiGraph(),
        )
        # Position and draw nodes
        fig, ax = plt.subplots(1, 1, figsize=size)
        sample_nodes = nodes[nodes["flags"] == 1].index
        other_nodes = nodes[nodes["flags"] == 0].index
        A = nx.nx_agraph.to_agraph(G)
        with A.subgraph(name="samples", rank="max") as s:
            s.add_nodes_from(sample_nodes)
        A.layout(prog="dot")

        pos = {
            n: [float(x) for x in A.get_node(n).attr["pos"].split(",")]
            for n in G.nodes()
        }

        nx.draw_networkx_nodes(
            G, pos, nodelist=sample_nodes, ax=ax, node_size=500, node_shape="s"
        )
        nx.draw_networkx_nodes(G, pos, nodelist=other_nodes, ax=ax, node_size=500)
        nx.draw_networkx_labels(G, pos, ax=ax)

        # Draw edges
        for (parent, child), edge_data in itertools.groupby(
            G.edges(data=True), key=lambda x: (x[0], x[1])
        ):
            edge_data = list(edge_data)
            curves = [0] if len(edge_data) == 1 else np.linspace(-1, 1, len(edge_data))
            for edge, curve in zip(edge_data, curves):
                edge_attributes = edge[2]
                ax.annotate(
                    "",
                    xy=pos[parent],
                    xycoords="data",
                    xytext=pos[child],
                    textcoords="data",
                    arrowprops=dict(
                        arrowstyle="-",
                        color="0.35",
                        shrinkA=10,
                        shrinkB=10,
                        patchA=None,
                        patchB=None,
                        connectionstyle=f"arc3,rad={0.2*curve}",
                    ),
                )
                label_offset = curve * 10
                ax.text(
                    (pos[parent][0] - pos[child][0]) * 0.3
                    + label_offset
                    + pos[child][0],
                    (pos[parent][1] - pos[child][1]) * 0.3 + pos[child][1],
                    edge_attributes["child_interval"],
                    ha="center",
                    bbox=dict(facecolor="white", edgecolor="none", pad=0.1),
                    fontsize=fontsize,
                    color="0.35",
                )
                ax.text(
                    (pos[parent][0] - pos[child][0]) * 0.7
                    + label_offset
                    + pos[child][0],
                    (pos[parent][1] - pos[child][1]) * 0.7 + pos[child][1],
                    edge_attributes["parent_interval"],
                    ha="center",
                    bbox=dict(facecolor="white", edgecolor="none", pad=0.1),
                    fontsize=fontsize,
                    color="0.35",
                )

        return fig, ax, G, pos, sample_nodes

    def plot_gig_traversal(self, u, x, size=(10, 10), fontsize=8):
        """
        Plot GIG traversal algorithm with slider, showing how the local tree
        is constructed in the process
        """
        # Base GIG plot
        fig, ax, G, pos, sample_nodes = self.plot_gig(size=size, fontsize=fontsize)
        plt.close()

        # Capture the data
        traversal_steps = []

        def local_tree_callback(**kwargs):
            traversal_steps.append(kwargs)

        self.local_tree_bottom_up(u, x, callback=local_tree_callback)

        # Construct the visualisation
        step_slider = IntSlider(
            value=0, min=0, max=len(traversal_steps) - 1, step=1, description="Step"
        )
        gig_output = Output()
        tree_output = Output()
        highlight_elements = []

        def on_step_change(change):
            step = change["new"]
            tables = traversal_steps[step]["tables"]
            kwargs = {k: v for k, v in traversal_steps[step].items() if k != "tables"}

            with gig_output:
                clear_output(wait=True)
                for artist in highlight_elements:
                    artist.remove()
                highlight_elements.clear()
                new_elements = plot_gig_highlight(
                    **kwargs,
                    ax=ax,
                    G=G,
                    pos=pos,
                    sample_nodes=sample_nodes,
                    fontsize=fontsize,
                )
                if new_elements:
                    highlight_elements.extend(new_elements)
                display(fig)

            with tree_output:
                clear_output(wait=True)
                svg_tree = plot_tree(u, tables, size=(400, 400))
                display(SVG(svg_tree))

        step_slider.observe(on_step_change, names="value")

        # Construct the buttons
        left_button = Button(description="<", tooltip="Previous Step")
        right_button = Button(description=">", tooltip="Next Step")

        def on_left_button_clicked(b):
            if step_slider.value > 0:
                step_slider.value -= 1

        def on_right_button_clicked(b):
            if step_slider.value < step_slider.max:
                step_slider.value += 1

        left_button.on_click(on_left_button_clicked)
        right_button.on_click(on_right_button_clicked)

        display(
            VBox(
                [
                    HBox([gig_output, tree_output]),
                    HBox([left_button, step_slider, right_button]),
                ]
            )
        )
        on_step_change({"new": 0})


def plot_gig_highlight(
    edge,
    matching_edges,
    direction,
    x_old,
    x_new,
    ax,
    G,
    pos,
    sample_nodes,
    fontsize=8,
    color="black",
):
    """
    Given various data from a GIG plot, return graphics elements which
    highlight a particular edge and associated x values.
    """
    parent = edge.parent
    child = edge.child
    id = edge.id
    node_shape = "o"

    if direction == "up":
        source = child
        destination = parent
        arrowstyle = "->"
    else:
        source = parent
        destination = child
        arrowstyle = "<-"
        if destination in sample_nodes:
            node_shape = "s"

    curves = (
        [0] if len(matching_edges) == 1 else np.linspace(-1, 1, len(matching_edges))
    )
    matching_edges.reset_index(inplace=True, drop=True)
    curve_index = matching_edges[matching_edges["id"] == id].index[0]
    curve = curves[curve_index]
    edge_attributes = {}
    edge_attributes["parent_interval"] = f"[{edge.parent_left}, {edge.parent_right})"
    edge_attributes["child_interval"] = f"[{edge.child_left}, {edge.child_right})"
    label_offset = curve * 10

    new_elements = []

    # Add node highlight
    highlighted_node = nx.draw_networkx_nodes(
        G,
        pos,
        nodelist=[destination],
        ax=ax,
        node_size=500,
        node_color="red",
        node_shape=node_shape,
    )
    new_elements.append(highlighted_node)

    # Show x values
    text1 = ax.text(
        pos[source][0] + 11 + len(str(x_old)) * 3,
        pos[source][1],
        f"{x_old}",
        ha="center",
        va="center",
        fontsize=fontsize + 2,
        color=color,
    )
    text2 = ax.text(
        pos[destination][0] + 11 + len(str(x_new)) * 3,
        pos[destination][1],
        f"{x_new}",
        ha="center",
        va="center",
        fontsize=fontsize + 2,
        color=color,
    )
    new_elements.extend([text1, text2])

    highlighted_edge = ax.annotate(
        "",
        xy=pos[parent],
        xycoords="data",
        xytext=pos[child],
        textcoords="data",
        arrowprops=dict(
            arrowstyle=arrowstyle,
            color="black",
            shrinkA=10,
            shrinkB=10,
            patchA=None,
            patchB=None,
            connectionstyle=f"arc3,rad={0.2*curve}",
            linewidth=2,
        ),
    )
    new_elements.append(highlighted_edge)

    # Add edge label texts
    label1 = ax.text(
        (pos[parent][0] - pos[child][0]) * 0.3 + label_offset + pos[child][0],
        (pos[parent][1] - pos[child][1]) * 0.3 + pos[child][1],
        edge_attributes["child_interval"],
        ha="center",
        bbox=dict(facecolor="white", edgecolor="none", pad=0.1),
        fontsize=fontsize,
        color=color,
    )
    label2 = ax.text(
        (pos[parent][0] - pos[child][0]) * 0.7 + label_offset + pos[child][0],
        (pos[parent][1] - pos[child][1]) * 0.7 + pos[child][1],
        edge_attributes["parent_interval"],
        ha="center",
        bbox=dict(facecolor="white", edgecolor="none", pad=0.1),
        fontsize=fontsize,
        color=color,
    )
    new_elements.extend([label1, label2])

    return new_elements


def plot_tree(u, tables, size=(400, 400)):
    tables.sort()
    ts = tables.tree_sequence()
    mut_labels = {}
    for mut in ts.mutations():
        mut_labels[mut.id] = ",".join(str(val) for val in mut.metadata["omega"])
    style = (
        ".node > .lab {font-size: 18px; font-family: Arial}"
        ".mut > .lab {font-size: 14px; font-family: Arial}"
        f".n{u} .sym {{fill: blue}}"
        f".node.n{u} > .lab {{fill: blue}}"
    )
    plot = ts.first().draw_svg(mutation_labels=mut_labels, size=size, style=style)
    return plot


gig3 = GIG(table_group3)
ts = gig3.local_tree_tskit(u=0, x=0)
```
